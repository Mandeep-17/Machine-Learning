# Convert Genres

```python
m = movies['genres'][0]
m
```

```python
ast.literal_eval(m)
```

```python
def  convert_genres(item):
    temp_lst = []
    item = ast.literal_eval(item)
    for dct in item:
        temp_lst.append(dct['name'])
    return temp_lst    
```

```python
convert_genres(m)
```

```python
movies['genres'] = movies['genres'].apply(convert_genres)
```

# Convert Cast

```python
c =movies['cast'][0]
```

```python
ast.literal_eval(c)
```

```python
def convert_cast(item):
    temp_lst = []
    item = ast.literal_eval(item)
    count = 3
    for dct in item:
        if count<=3:
            temp_lst.append(dct['name'])
        else:
            break    
    return temp_lst    
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

# Convert  Keyword

```python
k = movies['keywords'][0]
```

```python
ast.literal_eval(k)
```

```python
def convert_keyword(items):
    temp_lst = []
    item = ast.literal_eval(items)
    count =1
    for dct in item:
        if count<=6:
            temp_lst.append(dct['name'])
        else:
            break
    return temp_lst    
```

```python
movies['keywords'] = movies['keywords'].apply(convert_keyword)
```

# Convert  Crew

```python
 cr = movies['crew'][0]
```

```python
ast.literal_eval(cr)
```

```python
def convert_crew(items):
    temp_lst = []
    item = ast.literal_eval(items)
    for dct in item:
        if dct['job']=='Director':
            temp_lst.append(dct['name'])
    return temp_lst    
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

# Remove Space

```python
def remove_space(item):
    temp_lst = []
    for i in item:
        temp_lst.append(''.join(i.split()))
    return temp_lst    
```

```python
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
movies['genres'] = movies['genres'].apply(remove_space)
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
```

```python
movies['crew'] = movies['crew'].apply(remove_space)
```

# Use of Porterstemmer

```python
import nltk
```

```python
from nltk.stem import PorterStemmer

stemming = PorterStemmer()
```

```python
def stemming_words(items):
    temp_lst = []
    for word in items:
        temp_lst.append(stemming.stem(word))
    temp_lst = list(set(temp_lst)) # removes duplicate values
    return temp_lst
```

```python
 movies_2['Tag'] = movies_2['Tag'].apply(stemming_words)
```

```python
movies_2.head()
```

```python
movies_2['Tag'] = movies_2['Tag'].apply(lambda item:' '.join(item))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit(movies_2['Tag'])
```

```python
movie_metrix = cv.transform(movies_2['Tag']).toarray()
```

```python
movie_metrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

sm = cosine_similarity(movie_metrix)
```

```python
sm[0]
```

```python
sorted(sm[0],reverse=True)
```

```python
a = list(enumerate(sm[0]))
```

```python
sorted(a,reverse=True,key = lambda x: x[1])
```

```python
def recommend(movie):
    index = movies_2[movies_2['title'] == movie].index[0]
    distances = sorted(list(enumerate(sm[index])),reverse=True,key = lambda x : x[1])
    for i in distances[1:6]:
        print(movies_2.iloc[i[0]]['title'])
```

```python
recommend("Avatar")
```
